```python
from pyspark.sql import SparkSession
spark = (
    SparkSession.builder
    .appName("spark-iceberg-gravitino-test")

    .config(
        "spark.jars",
        "/opt/extra-jars/iceberg-spark-runtime-3.5_2.12-1.5.2.jar"
    )

    .config(
        "spark.sql.extensions",
        "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions"
    )

    .config("spark.sql.catalog.grav", "org.apache.iceberg.spark.SparkCatalog")
    .config("spark.sql.catalog.grav.catalog-impl",
            "org.apache.iceberg.rest.RESTCatalog")
    .config("spark.sql.catalog.grav.uri",
            "http://iceberg-sandbox-gravitino:9001/iceberg")

    # ‚ùå REMOVE THIS LINE
    # .config("spark.sql.catalog.grav.warehouse", "s3a://iceberg/")

    .getOrCreate()
)
```
